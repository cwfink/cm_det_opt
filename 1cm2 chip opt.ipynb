### 4% Surface coverage device

```python

desired_coverage = 0.04

l_tes0 = 100e-6
l_tes_bnds = [5e-6, 300e-6]
l_overlap0 = 10e-6
l_overlap_bnds = [5e-06, 40e-6]
l_fin0 = 100e-6
l_fin_bnds = [10e-6, 500e-6]
n_fin0 = 4
n_fin_bnds = [2,6]

res_n = 250e-3
rp = 2e-3
ahole = 4.9e-11

bounds=[l_tes_bnds,l_overlap_bnds ,l_fin_bnds , n_fin_bnds]
fixrn=True

det_opt_4, eres, params = optimize.optimize_detector( tes_length0=l_tes0,
                                                    tes_l_overlap0=l_overlap0,
                                                    l_fin0=l_fin0,
                                                    n_fin0=n_fin0,
                                                    per_Al = desired_coverage,   
                                                    fixrn=fixrn,
                                                    #tes_width=w,
                                                    abs_type="Si", 
                                                    abs_shape="square", 
                                                    abs_height=1e-3, 
                                                    abs_width=1e-2, 
                                                    w_safety=3e-3,
                                                    sigma=220000000.0,
                                                    rn=res_n,
                                                    rp=rp,
                                                    L_tot=5e-08,
                                                    ahole=ahole, 
                                                    con_type='ellipse',
                                                    veff_WAloverlap=.45,
                                                    bounds = bounds)
```

### 0.25% Surface coverage device


```python

desired_coverage = .0025
res_n = 500e-3
res_n = 500e-3

det_opt_01, eres, params = optimize.optimize_detector( tes_length0=l_tes0,
                                                    tes_l_overlap0=l_overlap0,
                                                    l_fin0=l_fin0,
                                                    n_fin0=n_fin0,
                                                    per_Al = desired_coverage,                               
                                                    #tes_width=w,
                                                    abs_type="Si", 
                                                    abs_shape="square", 
                                                    abs_height=1e-3, 
                                                    abs_width=1e-2, 
                                                    w_safety=3e-3,
                                                    sigma=220000000.0,
                                                    rn=res_n,
                                                    rp=rp,
                                                    L_tot=5e-08,
                                                    ahole=ahole, 
                                                    con_type='ellipse')
#det_opt_01.QET.TES.print()
```

```python
n = 30
ltes = np.linspace(20e-6, 100e-6, n)
lf = np.linspace(40e-6,150e-6, n)

plot.plot_ltes_vs_lfin(ltes, lf, det_opt_01, val='energy')
plot.plot_ltes_vs_lfin(ltes, lf, det_opt_01, val='al')
```

```python
det = det_opt_01
absorber = det._absorber
tes = det.QET.TES
qet = det.QET

rn = 1

abso1 = dark.Absorber(name=absorber._name, shape=absorber._shape,
                    height=absorber._h, width=absorber._width,
                    w_safety=absorber._w_safety)
tes1 = dark.TES(length=30e-6, width=tes.w, l_overlap=tes.l_overlap, n_fin=tes.n_fin, sigma=tes.sigma,
         rn=rn, rsh=tes.rsh, rp=tes.rp, L_tot=tes.L, tload=tes.tload,
         h=tes.h, veff_WAloverlap=tes.veff_WAloverlap, veff_WFinCon=tes.veff_WFinCon, 
         con_type=tes.con_type, material=tes.material, operating_point=tes.fOp,
         alpha=tes.alpha, beta=tes.beta, n=tes.n, Qp=tes.Qp, t_mc=tes.t_mc)
qet1 = dark.QET(l_fin=80e-6, h_fin=qet.h_fin, TES=tes1, ahole=qet.ahole, ePQP=qet.ePQP,
           eff_absb=qet.eff_absb, wempty=qet.wempty, wempty_tes=qet.wempty_tes, 
           type_qp_eff=qet.type_qp_eff)
det1 = dark.Detector(abso1, qet1, n_channel=det._n_channel, 
                freqs=det.freqs)

det1.calc_res(), det1._fSA_qpabsorb*100
```

```python
n = 30
ltes = np.linspace(10e-6, 50e-6, n)
lf = np.linspace(40e-6,150e-6, n)

plot.plot_ltes_vs_lfin(ltes, lf, det1, val='energy')
r = plot.plot_ltes_vs_lfin(ltes, lf, det1, val='al')
```

```python
# Set some initial parameters
l = 140e-6 
w = 3.5e-06
l_overlap = 10e-6
n_fins = 6
sigma = 220000000.0
t0 = 70e-3 #41.5e-3
res_n = 90e-3
#con_type= 'ellipse'
con_type= 'modern'
l_f = 200e-6
h_fin = 600e-6
ahole = 4.9e-11

# create absorber 
absorber = dark.Absorber(name="Si", 
                         shape= "cylinder",
                         height= 1e-3, 
                         width=7.62e-2,  
                         w_safety=3e-3)
# create tes
tes = dark.TES(length=l,
       width=w, 
       l_overlap = l_overlap,
       n_fin=n_fins,
       sigma=sigma,
       rn=res_n,
       rsh = 5e-3,
       rp = 9e-3,   
   L_tot= 19e-08, material=materials.TESMaterial(Tc=t0))

# create qet
qet = dark.QET(l_fin=l_f, 
               h_fin=h_fin, 
               TES=tes, 
               ahole=ahole, 
               type_qp_eff=0)

# create detector
det = dark.Detector(absorber, qet)
print(det.calc_res())
det.print()
```
